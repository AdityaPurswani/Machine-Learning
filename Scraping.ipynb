```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/quotes/tag/{}"
url.format('love')
```

```python
res = requests.get(url)
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text)
```

```python
#print(soup.get_text())
```

```python
links = soup.find_all("a")
```
