# line fitting custom function

````python
def estimated_coef(x, y):
  n = np.size(x)

  mean_x = np.mean(x)
  mean_y = np.mean(y)

  SS_xy= np.sum(y*x) - n*mean_x*mean_y
  SS_xx = np.sum(x*x) - n*mean_x*mean_x

  # calculating regretion coefficients
  b_1 = SS_xy/SS_xx
  b_0 = mean_y - b_1*mean_x

  return b_0, b_1


def plot_regretion(x, y, b):
  plt.scatter(x, y, cmap= 'gray',color='r', marker='x', s = 60)
  y_pred = b[0]+b[1]*x
  plt.plot(x, y_pred, color='g')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()
````

````python

def main():
  a = np.random.randint(1, 1000, 10)
  c = np.random.randint(1, 500, 10)
  b = estimated_coef(a, c)
  print("Estimated coef are", b[0], b[1])
  plot_regretion(a, c, b)

if __name__ == "__main__":
  main()
````

````python
def coef(x, y):
  n = np.size(x)

  m_x = np.mean(x)
  m_y = np.mean(y)

  SS_xy = np.sum(y*x) - n*m_y*m_x
  SS_xx = np.sum(x*x) - n*m_x*m_x

  b1 = SS_xy/SS_xx
  b0 = m_y - b1*m_x

  return (b0, b1)

def plot(x, y, b):
  plt.scatter(x, y, cmap='gray', marker='o')
  y_pred = b[0] + b[1]*x
  plt.plot(x, y_pred)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.show()

def main():
  x = np.random.randint(1,20,5)
  y = np.random.randint(3, 50, 5)
  b = coef(x, y)
  print("Estimated coef", b[0], b[1])
  plot(x, y,b)

if __name__ == "__main__":
  main()
````
