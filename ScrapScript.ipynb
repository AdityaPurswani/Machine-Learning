```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/quotes/tag/{}?page={}"
```

```python
def get_quotes(url):    
    res = requests.get(url)
    quotes = []
    soup = BeautifulSoup(res.text)
    quote_class = soup.find_all("div", attrs= {"class": "quote"})
    for quote_div in quote_class:
        quoteText_div = quote_div.find_next("div", attrs={"class": "quoteText"})
        strip = quoteText_div.text.strip()
        strip_ls = strip.split("\n")
        quote = strip_ls[0][1:-1]
        author = strip_ls[-1].strip()
        left_div = quote_div.find_next("div", attrs={"class": "left"})

        tags = [tag.text for tag in left_div.find_all("a")]
        quote_item ={"text": quote, "author": author, "tags":tags}
        quotes.append(quote_item)
    return quotes
```

```python
total = []
for i in range(1, 6):
    total.extend(get_quotes(url.format("kindness", i)))
    print("Processed page", i)
```
