```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("/content/cars.csv")
```

```python
df
```

```python
from sklearn import linear_model
X= df[["Weight", "Volume"]]
y = df["CO2"]

model = linear_model.LinearRegression()
model.fit(X, y)
predictedCO2 = model.predict([[2300, 1300]])
print(predictedCO2)
```

```python
print(model.coef_)
```
